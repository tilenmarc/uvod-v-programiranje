```python
filmi = pd.read_csv("filmi.csv", index_col="id")
```

```python
filmi.head(10)
```

```python
filmi.tail(5)
```

```python
filmi["profit"] = filmi["zasluzek"] - filmi["cena"]
```

```python
filmi.tail(10)
```

```python
filmi[["naslov", "cena"]]
```

```python
najvecji_zasluzkarji = filmi[filmi["zasluzek"] > 10**9]
```

```python
najvecji_zasluzkarji
```

```python
filmi["desetletje"] = filmi["leto"] - (filmi["leto"] % 10) 
```

```python
filmi.sort_values("zasluzek", ascending=False).head(10)
```

```python
ocena_po_desetletjih = filmi.groupby("desetletje")["ocena"].mean()
```

```python
filmi.groupby("desetletje")["naslov"].count()
```

```python
ocena_po_desetletjih.plot()
```

```python
ocena_po_desetletjih.plot.bar()
```

```python
najvecji_zasluzkarji[["naslov", "ocena"]].sort_values("ocena").plot.bar(x="naslov", y="ocena")
```

```python
filmi[filmi["zasluzek"] > 5*10**8].plot.scatter(x="zasluzek", y="ocena")
```

```python
filmi.plot.scatter(x="cas", y="ocena")
```

```python
filmi["cas_zaokrozen"] = (filmi["cas"] // 10) * 10
```

```python
filmi.groupby("cas_zaokrozen")["ocena"].mean().plot.bar()
```

```python
osebe = pd.read_csv("osebe.csv", index_col="id")
osebe_filmi = pd.read_csv("osebe_filmi.csv")
zanri = pd.read_csv("filmi_zanri.csv")
```

```python
filmi_z_zanri = pd.merge(filmi, zanri, left_on="id", right_on="id_filma")
```

```python
komedije = filmi_z_zanri[filmi_z_zanri["zanr"] == "comedy"]
komedije[komedije["desetletje"] == 2000]
```

```python
filmi_z_zanri.groupby("zanr")["cas"].mean().sort_values().plot.bar()
```

```python
vloge = pd.merge(osebe, osebe_filmi, left_on="id", right_on="id_igralca")
vloge.groupby("ime")["id_filma"].count().sort_values(ascending=False)
```
