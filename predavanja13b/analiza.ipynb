```python
zanri_filmi = pd.merge(filmi, zanri, left_on="id", right_on="id_filma")
```

```python
romance_komedije = zanri_filmi[(zanri_filmi["zanr"] == "comedy") | (zanri_filmi["zanr"] == "romance")].groupby(["id_filma", "naslov"]).size()
```

```python
import networkx as nx
```

```python
G = nx.from_pandas_edgelist(osebe_filmi.head(1000), "id_osebe", "id_filma")
G = nx.subgraph(G, list(nx.connected_components(G))[0])
```

```python
import math
d = dict(G.degree)
nx.draw(G, node_size=[math.log(v) * 20 for v in d.values()])
```
